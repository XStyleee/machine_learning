10) Создать матрицу случайных чисел размера 10 на 10 и отнять среднее по строкам из каждого элемента строки

```python
import numpy as np
matrix = np.random.rand(10, 10)
row_means = matrix.mean(axis=1, keepdims=True)
adjusted_matrix = matrix - row_means
matrix, adjusted_matrix
```

12) Создать матрицу случайных чисел 5x5. Добавить к ней столбецы максимумов и минимумов по строкам. Получится матрица 5x7.  

```python
matrix = np.random.rand(5, 5)
row_max = matrix.max(axis=1, keepdims=True)
row_min = matrix.min(axis=1, keepdims=True)
expanded_matrix = np.hstack((matrix, row_max, row_min))
matrix, expanded_matrix
```

13) Вектор чисел от 1 до 51 преобразовать в вектор нечетных чисел и в вектор четных чисел, поменять их размерности, чтобы получились две матрица. Найти произведение этих матриц. Найти обратную матрицу эётого произведения.  

```python
import numpy as np

vector = np.arange(1, 52)

odd_numbers = vector[vector % 2 != 0]
even_numbers = vector[vector % 2 == 0]


odd_matrix = odd_numbers[:25].reshape(5, 5)


even_matrix = even_numbers[:25].reshape(5, 5)


product_matrix = np.dot(odd_matrix, even_matrix)


det = np.linalg.det(product_matrix)

if det != 0:
    inverse_matrix = np.linalg.inv(product_matrix)
else:
    inverse_matrix = "Обратной матрицы не существует, так как определитель равен нулю."

odd_matrix, even_matrix, product_matrix, det, inverse_matrix
```
